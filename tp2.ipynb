## Exercício 1: Implementação de Trie com Inserção e Fim de Palavra

### Descrição Técnica da Trie
A árvore Trie é uma estrutura de caminhos onde cada nó representa um caractere. O caminho da raiz até um nó com a marcação `is_end = True` representa uma palavra válida armazenada. Esta estrutura é altamente eficiente para verificar prefixos comuns em tempo proporcional ao tamanho da palavra pesquisada.

```python
# Exercício 1: Classe TrieNode
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

# Exercício 1: Classe Trie
class Trie:
    def __init__(self):
        self.root = TrieNode()

    # Exercício 1: Método inserção
    def insert(self, word):
        if word == "":
            return
        atual = self.root
        for char in word:
            if char not in atual.children:
                atual.children[char] = TrieNode()
            atual = atual.children[char]
        atual.is_end = True

# Exercício 1: Testes mínimos
trie_teste = Trie()
print("Inserindo palavra nova 'car':")
trie_teste.insert("car")
print("Inserindo palavra repetida 'car':")
trie_teste.insert("car")
print("Inserindo prefixo 'cart' e 'carro':")
trie_teste.insert("cart")
trie_teste.insert("carro")
```

## Exercício 2: Busca Exata em Trie (search)

### Diferença entre Prefixo e Palavra Completa
- **Prefixo**: Existe como caminho na árvore (`children`), mas o nó final possui `is_end = False`.
- **Palavra Completa**: Existe como caminho na árvore e o nó final possui `is_end = True`.

### Decisão sobre Palavra Vazia
Optou-se por retornar `False` para a busca de palavra vazia `""`, pois não há representação de caracteres vazios na árvore Trie e a raiz não é marcada como fim de palavra por padrão.

```python
# Exercício 2: Busca exata
def search(self, word):
    if word == "":
        return False
    atual = self.root
    for char in word:
        if char not in atual.children:
            return False
        atual = atual.children[char]
    return atual.is_end

Trie.search = search

# Exercício 2: Testes
print(f"Palavra 'car' existe? {trie_teste.search('car')}")
print(f"Palavra 'carro' existe? {trie_teste.search('carro')}")
print(f"Prefixo 'ca' existe como palavra? {trie_teste.search('ca')}")
print(f"Palavra inexistente 'house' existe? {trie_teste.search('house')}")
print(f"Palavra vazia existe? {trie_teste.search('')}")
```

## Exercício 3: Busca por Prefixo (starts_with)

O método `starts_with` verifica apenas se o caminho correspondente aos caracteres existe na árvore, ignorando a propriedade `is_end`.

```python
# Exercício 3: Busca prefixo
def starts_with(self, prefix):
    if prefix == "":
        return False
    atual = self.root
    for char in prefix:
        if char not in atual.children:
            return False
        atual = atual.children[char]
    return True

Trie.starts_with = starts_with

# Exercício 3: Registro de testes
prefixos = ["car", "carr", "ca", "house", ""]
for pref in prefixos:
    print(f"Prefixo '{pref}' existe na Trie? {trie_teste.starts_with(pref)}")
```

## Exercício 4: Coleta de Palavras (Traversal)

A coleta recursiva permite recuperar de forma ordenada todas as palavras contidas em uma determinada subárvore a partir de um nó inicial, acumulando incrementalmente o prefixo.

```python
# Exercício 4: Coleta recursiva
def _collect_words(self, node, current_prefix, results):
    if node.is_end:
        results.append(current_prefix)
    for char, child_node in sorted(node.children.items()):
        self._collect_words(child_node, current_prefix + char, results)

# Exercício 4: Coleta pública
def collect_words(self, node, prefix):
    results = []
    self._collect_words(node, prefix, results)
    return results

Trie._collect_words = _collect_words
Trie.collect_words = collect_words

# Exercício 4: Testes e Prova de Integridade
todas_palavras = trie_teste.collect_words(trie_teste.root, "")
print(f"Todas as palavras a partir da raiz: {todas_palavras}")
sub_palavras = trie_teste.collect_words(trie_teste.root.children["c"].children["a"].children["r"], "car")
print(f"Palavras a partir do prefixo 'car': {sub_palavras}")
```

## Exercício 5: Autocomplete Lexicográfico

### Análise de Complexidade Temporal (Pior Caso)
- **Tamanho do prefixo ($P$)**: O tempo para encontrar o nó do prefixo é $O(P)$.
- **Subárvore**: O tempo para percorrer recursivamente e coletar todas as palavras é $O(V_{sub})$, onde $V_{sub}$ representa o número de nós na subárvore.
- **Ordenação**: A ordenação das palavras coletadas leva $O(N_{words} \log N_{words} \times L)$, onde $N_{words}$ é o número de palavras encontradas e $L$ é o comprimento máximo de uma palavra.
- **Complexidade Total**: $O(P + V_{sub} + N_{words} \log N_{words} \times L)$.

```python
# Exercício 5: Método autocomplete
def autocomplete(self, prefix, k):
    if prefix == "":
        return []
    atual = self.root
    for char in prefix:
        if char not in atual.children:
            return []
        atual = atual.children[char]
    candidates = self.collect_words(atual, prefix)
    candidates.sort()
    return candidates[:k]

Trie.autocomplete = autocomplete

# Inserindo mais palavras para testes
trie_teste.insert("casa")
trie_teste.insert("cachorro")
trie_teste.insert("carpa")

print(f"Autocomplete para 'ca' (k=2): {trie_teste.autocomplete('ca', 2)}")
print(f"Autocomplete para 'ca' (k=5): {trie_teste.autocomplete('ca', 5)}")
print(f"Autocomplete para 'house' (k=3): {trie_teste.autocomplete('house', 3)}")
```

## Exercício 7: Representação de Grafo - Lista de Adjacência

A lista de adjacência mapeia cada vértice para a coleção de seus vizinhos. É altamente eficiente no uso de memória para representar grafos esparsos (onde a quantidade de arestas é muito menor que o quadrado do número de vértices).

```python
# Exercício 7: Grafo adjacência
class GraphAdjList:
    def __init__(self):
        self.adj = {}

    # Exercício 7: Vértice lista
    def add_vertex(self, v):
        if v not in self.adj:
            self.adj[v] = set()

    # Exercício 7: Aresta lista
    def add_edge(self, u, v, directed=False):
        self.add_vertex(u)
        self.add_vertex(v)
        self.adj[u].add(v)
        if not directed:
            self.adj[v].add(u)

    # Exercício 9: Consulta lista
    def has_edge(self, u, v):
        if u not in self.adj:
            return False
        return v in self.adj[u]

# Exercício 7: Criando o grafo de teste
grafo_lista = GraphAdjList()
vertices = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
for vertice in vertices:
    grafo_lista.add_vertex(vertice)

arestas = [
    ("A", "B"), ("A", "C"), ("B", "D"), ("C", "D"), ("D", "E"),
    ("E", "F"), ("F", "G"), ("G", "H"), ("H", "I"), ("I", "J"),
    ("E", "H"), ("B", "F")
]
for u, v in arestas:
    grafo_lista.add_edge(u, v)

# Exibição da Lista de Adjacência
for u in sorted(grafo_lista.adj.keys()):
    print(f"Vértice {u}: {sorted(list(grafo_lista.adj[u]))}")
```

## Exercício 8: Representação de Grafo - Matriz de Adjacência

A matriz de adjacência utiliza uma matriz bidimensional para marcar a conectividade entre os índices dos vértices.

```python
# Exercício 8: Grafo matriz
class GraphAdjMatrix:
    def __init__(self):
        self.index = {}
        self.mat = []
        self.vertices = []

    # Exercício 8: Vértice matriz
    def add_vertex(self, v):
        if v not in self.index:
            self.index[v] = len(self.vertices)
            self.vertices.append(v)
            for row in self.mat:
                row.append(0)
            self.mat.append([0] * len(self.vertices))

    # Exercício 8: Aresta matriz
    def add_edge(self, u, v, directed=False):
        self.add_vertex(u)
        self.add_vertex(v)
        idx_u = self.index[u]
        idx_v = self.index[v]
        self.mat[idx_u][idx_v] = 1
        if not directed:
            self.mat[idx_v][idx_u] = 1

    # Exercício 8: Consulta matriz
    def has_edge(self, u, v):
        if u not in self.index or v not in self.index:
            return False
        idx_u = self.index[u]
        idx_v = self.index[v]
        return self.mat[idx_u][idx_v] == 1

# Exercício 8: Criando grafo correspondente
grafo_matriz = GraphAdjMatrix()
for vertice in vertices:
    grafo_matriz.add_vertex(vertice)
for u, v in arestas:
    grafo_matriz.add_edge(u, v)

# Exibição da Matriz
print("Índices dos vértices na matriz:", grafo_matriz.index)
print("Matriz de Adjacência:")
for linha in grafo_matriz.mat:
    print(linha)
```

## Exercício 9: Definições Operacionais

### 1. Vértice
- **Definição**: Elemento ou nó do grafo que pode conter informações ou identificadores textuais.
- **Implementação**: Representado pelas chaves no dicionário `self.adj` da lista ou pela lista `self.vertices` na matriz.
- **Exemplo**: O vértice `"A"` está armazenado no dicionário `grafo_lista.adj["A"]`.

### 2. Aresta
- **Definição**: Conexão entre dois vértices, podendo ser direcionada ou não.
- **Implementação**: Representado pela inclusão do vértice adjacente na lista/conjunto de vizinhos ou pelo valor `1` na matriz.
- **Exemplo**: A aresta `"A" - "B"` está registrada em `grafo_lista.adj["A"]` (que contém `"B"`) e `grafo_lista.adj["B"]` (que contém `"A"`).

### 3. Lista de Adjacência
- **Definição**: Estrutura de dados que armazena, para cada vértice, uma coleção dinâmica dos seus vértices adjacentes.
- **Implementação**: Atributo `self.adj` (dicionário de conjuntos) em `GraphAdjList`.
- **Exemplo**: `{ "A": {"B", "C"}, "B": {"A", "D", "F"}, ... }`.

### 4. Matriz de Adjacência
- **Definição**: Estrutura matricial quadrada $V \times V$ onde a célula no índice $(i, j)$ possui valor $1$ se há uma aresta conectando os vértices correspondentes, ou $0$ caso contrário.
- **Implementação**: Atributo `self.mat` (lista de listas de inteiros) em `GraphAdjMatrix`.
- **Exemplo**: O cruzamento de linha `"A"` e coluna `"B"` contém o valor $1$.

### Análise de Custo da Consulta `has_edge`
- **Lista de Adjacência**: Requer busca linear no conjunto de adjacências de $u$. O custo é proporcional ao grau do vértice, ou seja, $O(\text{grau}(u))$.
- **Matriz de Adjacência**: O acesso é direto por meio dos índices mapeados na memória da tabela bidimensional. O custo é de tempo constante $O(1)$.

## Exercício 10: Exportação em Formato Mermaid

### Regra para Evitar Duplicatas
Para grafos não-direcionados, cada aresta gera caminhos do tipo $u \to v$ e $v 	o u$. Para evitar a impressão de linhas duplicadas no arquivo Mermaid, as arestas são ordenadas alfabeticamente e salvas em um conjunto `seen` de tuplas. Somente arestas inéditas são formatadas na string final.

```python
# Exercício 10: Exportação Mermaid
def to_mermaid(self, directed=False):
    lines = ["graph TD" if directed else "graph LR"]
    seen = set()
    for u in sorted(self.adj.keys()):
        for v in sorted(self.adj[u]):
            if directed:
                lines.append(f"    {u} --> {v}")
            else:
                aresta = tuple(sorted([u, v]))
                if aresta not in seen:
                    seen.add(aresta)
                    lines.append(f"    {u} --- {v}")
    return "\n".join(lines)

GraphAdjList.to_mermaid = to_mermaid

# Exercício 10: Geração do bloco Mermaid
mermaid_output = grafo_lista.to_mermaid(directed=False)
print("Saída do Diagrama Mermaid:")
print(mermaid_output)
```

## Exercício 12: Decisões de Projeto e Complexidade

### Tabela de Operações da Trie

| Operação | Complexidade Temporal (Pior Caso) | Justificativa |
| :--- | :--- | :--- |
| **`insert(word)`** | $O(L)$ | Varre cada caractere da palavra de comprimento $L$. |
| **`search(word)`** | $O(L)$ | Navega ao longo do caminho de comprimento $L$ na árvore. |
| **`starts_with(prefix)`** | $O(P)$ | Navega ao longo do caminho de comprimento $P$ do prefixo. |
| **`autocomplete(prefix, k)`**| $O(P + V_{sub} + N \log N \times L)$ | Localiza o nó, coleta termos na subárvore e ordena lexicograficamente. |

### Tabela Comparativa de Representação de Grafos

| Operação / Critério | Lista de Adjacência | Matriz de Adjacência |
| :--- | :--- | :--- |
| **Inserção de Aresta** | $O(1)$ | $O(1)$ |
| **Consulta de Adjacência** | $O(\text{grau}(u))$ | $O(1)$ |
| **Consumo de Memória** | $O(V + E)$ | $O(V^2)$ |

### Justificativas de Escolha de Representação
1. **Grafo Esparso com milhões de vértices**: A escolha apropriada é a **Lista de Adjacência**. Como o número de arestas $E \ll V^2$, armazenar uma matriz quadrática consumiria muita memória desnecessariamente. A lista armazena apenas as conexões reais.
2. **Grafo Denso com poucos vértices**: A escolha ideal é a **Matriz de Adjacência**. Como o grafo possui muitas conexões ($E \approx V^2$), o overhead de ponteiros e estruturas dinâmicas das listas de adjacência é eliminado, e a consulta ocorre em tempo constante $O(1)$.

### Duas Decisões Técnicas Justificadas com Rigor
1. **Decisão na Trie (Dicionário de Filhos)**: Optou-se por utilizar dicionários dinâmicos (`dict`) no atributo `children` de cada nó. Essa decisão garante acesso rápido médio em $O(1)$ para cada caractere, reduzindo a complexidade temporal em relação a listas simples que exigiriam buscas lineares por caractere.
2. **Decisão no Grafo (Conjunto de Adjancência)**: A classe `GraphAdjList` armazena os vizinhos em uma estrutura de conjunto (`set`) do Python. Essa decisão permite a inserção e a verificação de duplicatas de arestas em tempo constante médio $O(1)$ em comparação a listas lineares que exigiriam tempo de varredura $O(\text{grau}(u))$.
